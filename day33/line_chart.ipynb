```python
import matplotlib.pyplot as plt
x=range(1,8)
y=[2,5,9,6,8,9,5]
plt.plot(x,y,color='green',marker='p',alpha=1,linestyle='-',linewidth=1)  # alpha透明度
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x=range(2,26,2)
y=[random.randint(15,30) for i in x]
plt.figure(figsize=(20,8),dpi=80)
print(list(x))
print(y)
x_ticks_lable=["{}:00".format(i) for i in x]
plt.xticks(x,x_ticks_lable,rotation=45)
y_ticks_lable=["{}℃".format(i) for i in range(min(y),max(y)+1)]
plt.yticks(range(min(y),max(y)+1),y_ticks_lable)
plt.plot(x,y) 
plt.show()
```
